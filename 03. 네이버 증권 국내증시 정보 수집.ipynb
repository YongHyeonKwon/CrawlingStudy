```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
# 1. 네이버 시가총액 사이트 정보 요청
res = req.get("http://finance.naver.com/sise/sise_market_sum.naver")
res.text
```

```python
# 2. 응답코드를 확인하고, 정상 응답이라면 bs통해서 데이터 변형=
soup = bs(res.text, "lxml")
```

```python
# 3. 변형된 데이터에서 종목명,현재가격 정보를 수집
jm = soup.select("#contentarea > div.box_type_l > table.type_2 > tbody > tr > td:nth-child(2) > a")
price = soup.select("#contentarea > div.box_type_l > table.type_2 > tbody > tr > td:nth-child(3)")
price
```

```python
len(price)
len(jm)
```

```python
# 4. 순수한 텍스트 정보로 변형작업(리스트를 제작)
jm_list = []
price_list = []
rank_list = []

for i in range(len(jm)) :
    jm_list.append(jm[i].text)
    price_list.append(price[i].text)
    rank_list.append(i+1)          
    
price_list
jm_list
rank_list
```

```python
# 5. 표를 제작, 한글이 깨지지 않게 csv파일로 저장
import pandas as pd
dic = {"종목명":jm_list, "가격" : price_list, "순위" : rank_list}

pinance = pd.DataFrame(dic)
```

```python
pinance.set_index("순위", inplace= True)
```

```python
pinance
```

```python
pinance.to_csv("국내증시.csv", encoding="euc-kr")
```
