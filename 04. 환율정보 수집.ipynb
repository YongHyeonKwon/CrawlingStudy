# iframe 데이터처리
- 화면상의 정보는 보이지만 실제 데이터는 다른 공간에 존재할 때 사용

```python
# iframe은 다른 공간에 존재하는 정보를 화면에 표출만 해준다.
# 해결법 = 실제 데이터가 존재하는 URL로 요청을 보내야한다,
# iframe태그의 src안에 있는 경로를 사용하면 된다,
# 대부분의 경로는 상대경로로 작성이 되어있다.
# 상대경로를 절대경로와 합쳐서 URL 제작
res = req.get("https://finance.naver.com/marketindex/exchangeList.naver")
soup = bs(res.text, "lxml")
```

```python
name = soup.select("body > div > table > tbody > tr > td.tit > a")
gijoon = soup.select("body > div > table > tbody > tr > td.sale")
name
```

```python
# 개행문자 혹은 공백 등을 지울 때는 strip() 활용
name_list = []
for i in name :
    name_list.append(i.text.strip())
name_list
```
