# Melon사이트 요청

```python
# response 406 => 클라이언트의 요청문제!
# Why? 브라우저가 아닌 코드로 접근해서 멜론서버에서 데이터를 리턴하지 않는다.
# How? 사람인척 속이는(브라우저를 동반해서 보내기)
import requests as req
req = req.get("https://www.melon.com/chart/index.htm") 
```

```python
# user-agent = 브라우저의 정보를 담고있는 데이터 
# 개발자도구 네트워크 탭에서 doc문서를 선택하고, headers의 맨 마지막에 존재 
head = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36"}
```

```python
import requests as req
res= req.get("https://www.melon.com/chart/index.htm", headers = head)
res.text
```

# BeautifulSoup 라이브러리
- req로 받아온 데이터를 컴퓨터가 이해할 수 있는 데이터로 가공

```python
# 멜론차트 시험에 나온다. 인코딩 빼먹으면 부분점수 차감, 
# breautifulsoup이 뭔가? 
```

```python
# stonesoup가 아닌 soup을 임포트!
# 수집한 데이터는 태그와 컨텐츠가 합쳐진 Eliments(요소)
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(res.text, "lxml")
```

```python
# select명령은 soup(가공된 html)에서 필요한 태그정보를 수집
news_title = soup.select("a.news_tit")
```

```python
# 순수한 텍스트(컨텐츠)정보만 저장 
text_list = []
for i in news_title :
    text_list.append(i.text)
```

```python
text_list
```

```python
import requests as req
from bs4 import BeautifulSoup as bs

res = req.get("https://finance.naver.com")
```

```python
soup = bs(res.text, "lxml")
chart = soup.select("#_topItems1 > tr > th > a")
```

```python
texts_list = []
for i in chart :
    texts_list.append(i.text)
```
