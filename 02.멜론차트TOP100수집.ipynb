```python
import requests as req
from bs4 import BeautifulSoup as bs 
```

```python
head = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36"}
res = req.get("https://www.melon.com/chart/index.htm", headers = head)
```

```python
soup = bs(res.text, "lxml")
```

```python
# 가수이름
# 클래스명 중간에 띄어쓰기가 있다는 것은 클래스명이 복수 개다! 
# 해결별 = 공백을 지우고 .으로 이어준다.
singer = soup.select("div.ellipsis.rank02 > a")
singer
```

```python
song = soup.select("div.ellipsis.rank01 > span > a")
song
```

```python
# 데이터의 개수가 정해진 데이터를 수집할 때는 
# 반드시 데이터의 개수를 체크!
len(singer) 
```

```python
len(song)
```

```python
# 부정확한 데이터가 수집되면
# 반드시 html코드정보를 다시 한 번 살펴봐야한다.
# 내가 수집할 요소가 각각 들고와 지면 한 번에 담고있는 부모를 수집
singer = soup.select("div.ellipsis.rank02 > span.checkEllipsis")
# select_one => 하나만 수집 데이터의 개수가 복수 개면 첫 번째 데이터
# select를 쓰면 데이터가 한 개여도 무조건 리스트 형태로 리턴
soup.select_one("div.ellipsis.rank02 > span.checkEllipsis")
```

```python
# 선택자 한 번에 복사해오는 방법
# html태그 > 우클릭 > copy > copy selector
# 주의점 = 무조건 선택한 태그 하나만 가지고 오는 선택자 
# 활용방법 = 복수 개를 가지고 오고 싶다면 nth-child를 지워주자!
```

```python
# 순수한 텍스트 정보만 담은 리스트 제작
# 변수 선언은 반복문 바로 위에 오면 좋다.
# 되도록이면 한셀에 모두 넣는게 좋음.

singer_list = []
song_list = []
rank_list = []
for i in range(len(singer)) :
    singer_list.append(singer[i].text)
    song_list.append(song[i].text)
    rank_list.append(i+1)
singer_list
song_list
rank_list
```

```python
import pandas as pd
```

```python
dic = {"가수이름":singer_list, "노래제목" : song_list, "순위" : rank_list}
```

```python
melon = pd.DataFrame(dic)
melon
```

```python
melon.set_index("순위", inplace= True)
```

```python
melon
```

```python
# encoding을 꼭 해야한다!
# 1. utf-8
# 2. euc-kr
# 3. utf-8-sig

melon.to_csv("멜론차트.csv", encoding="euc-kr")
```
